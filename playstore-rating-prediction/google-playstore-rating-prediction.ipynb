#### Music 이랑 Music & Audio랑 합칠 것인지...  -> 나중에

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
features = ['Category', 'Rating Count(k)',
       'Installs(k)', 'Free', 'Price', 'Size(M)',
       'Ad Supported', 'In App Purchases', 'Editors Choice']
x = data2[features]
y = data2['Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
```

```python
for i in range(5,20):
    model = KNeighborsRegressor(n_neighbors=i)
    model.fit(X_train, y_train)
    accuracy = model.score(X_test,y_test)
    print('neighbors',i,'Accuracy: ' + str(np.round(accuracy*100, 2)) + '%')
```

```python
model = KNeighborsRegressor(n_neighbors=11)
model.fit(X_train, y_train)
accuracy = model.score(X_test,y_test)
```

### 거리 척도의 단위 문제 - 표준화(유클리드 거리)

어쨋든 k-NN 알고리즘을 적용할 때는 데이터에 표준화가 필요하다는 것이다. 

```python
test = pd.read_csv('webapp_test_feature.csv')
```

```python
test.head()
```

```python
test = test[['App Name', 'Category', 'Rating Count', 'Installs', 'Free', 'Price', 'Size', 'Released', 'Last Updated', 'Ad Supported', 'In App Purchases', 'Editors Choice']]
values = {'Rating Count': test['Rating Count'].median()}
test = test.fillna(value=values)

def to_float(row):
    row = row.rstrip('+').replace(',','')
    return float(row)/1000

test['Installs'] = test['Installs'].apply(to_float)
test = test.rename(columns={'Installs':'Installs(k)'})
test['Size'][test['Size']=='Varies with device'] = np.nan

def k_to_m(row):
    row = str(row).replace(',','')
    if row[-1] == 'k':
        row = float(row[:-1])/1000
    elif row[-1] == 'M':
        row = row[:-1]
    return float(row)  
test['Size'] = test['Size'].apply(k_to_m)
test = test.rename(columns={'Size':'Size(M)'})
test['Size(M)'] = test['Size(M)'].fillna(test['Size(M)'].median())
category_li = test['Category'].unique().tolist()
category_dict = {}
for index, category in enumerate(category_li):
    category_dict[category] = index
def category_to_float(row):
    row = category_dict[row]
    return row
test['Category'] = test['Category'].apply(category_to_float)
def bool_to_int(row):
    if row:
        row = 1
    else:
        row = 0
    return row
test['Ad Supported'] = test['Ad Supported'].apply(bool_to_int)
test['In App Purchases'] = test['In App Purchases'].apply(bool_to_int)
test['Editors Choice'] = test['Editors Choice'].apply(bool_to_int)
test['Free'] = test['Free'].apply(bool_to_int)
test = test[['Category', 'Rating Count', 'Installs(k)', 'Free',
       'Price', 'Size(M)', 'Ad Supported',
       'In App Purchases', 'Editors Choice']]
```
