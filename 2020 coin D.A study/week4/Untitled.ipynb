```python
doc = pd.read_csv('shopping.csv', encoding='cp949')
doc.head()
```

```python
doc.shape
```

```python
doc['국가(대륙)별'].value_counts()
```

```python
usa = doc[doc['국가(대륙)별']=='미국']
usa
```

```python
usa_total = usa[(usa['상품군별']!='합계') & (usa['판매유형별']=='계')]
usa_total.head()
```

```python
usa_total.info()
```

```python
columns = list(usa_total.columns)
columns = columns[3:15]
columns
```

```python
for column in columns:
    usa_total[column] = usa_total[column].astype('int64')
usa_total.info()
```

```python
usa_total.head()
```

```python
books = usa_total[usa_total['상품군별']=='서 적']
books
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x="상품군별", y="2014 1/4", data=usa_total)
```
