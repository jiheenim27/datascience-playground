```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import os

if os.name == 'posix':
    plt.rc('font', family='AppleGithic')
else:
     plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
%config inlineBackend.figure_format = 'retina'
```

```python
doc = pd.read_csv('서울시설공단_서울도시고속도로 노선별 시간대별 속도(2018)..csv', encoding='cp949')
doc.head()
```

```python
doc.info()
```

```python
weekend = doc[doc['구분']=='주말']
weekend
```

```python
# 오전 6시부터 11시까지
```

```python
weekend.columns
```

```python
weekend_am = weekend[(weekend['시간대']<12) & (weekend['시간대']>5)]
weekend_am.head()
```

```python
weekdays = doc[doc['구분']=='평일']
```

## 주말 오전 고속도로별 교통통계 현황

```python
plt.figure(figsize=(12,12))
sns.pointplot(data=weekend_am, x='시간대', y='속도', hue='노선', ci=None)
```

```python
plt.figure(figsize=(12,12))
sns.lineplot(data=weekend_am, x='시간대', y='속도', hue='노선', ci=None) 
```

## 주말 고속도로별 교통통계 현황

```python
plt.figure(figsize=(12,12))
sns.lineplot(data=weekend_am, x='시간대', y='속도', hue='노선', ci=None) 
```

```python
plt.figure(figsize=(12,12))
sns.pointplot(data=weekend, x='시간대', y='속도', hue='노선', ci=None) 
```

```python
plt.figure(figsize=(12,12))
sns.lineplot(data=weekend, x='시간대', y='속도', hue='노선', ci=None) 
```
